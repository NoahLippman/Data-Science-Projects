Based on the Above data "average exit velocity","barrel rate", and "hard hit rate" are the greatest predictors of wOBA

```python
import matplotlib.pyplot as plt 
from mlxtend.plotting import scatterplotmatrix
cols = ["woba","avg_exit_velocity", "avg_launch_angle", "barrel_batted_rate", "hard_hit_percent",
        "z_swing_percent", "oz_swing_percent", "iz_contact_percent"]

scatterplotmatrix(numbersData[cols].values, figsize= (25,25),names = cols, alpha = .5)
plt.tight_layout()
plt.show()
```

```python
X = numbersData.drop("woba", axis = 1).values
Y = numbersData["woba"].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = .2,
                                                    random_state = 21)
```

```python
from sklearn.linear_model import Lasso
scores = []
alphas = [.00001, 1.0, 10.0, 20.0, 35.0, 50.0]
for alpha in alphas:
    lasso = Lasso(alpha = alpha)
    lasso.fit(X_train, y_train)
    lasso_pred = lasso.predict(X_test)
    score = lasso.score(X_test, y_test)
    scores.append(score)
scores
```

```python
names = numbersData.drop("woba", axis = 1).columns
lasso = Lasso(alpha = .00001)
lasso_coef = lasso.fit(X,Y).coef_
plt.bar(names, lasso_coef)
plt.xticks(rotation = 45)
plt.show
```

```python
lWOBA = [i for i in lasso.predict(X)]
data["lWOBA"] = lWOBA
```

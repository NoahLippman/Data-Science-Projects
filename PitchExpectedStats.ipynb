The Nan is due to the formatting from Tableu.
To deal with this, we have to replace the NANs with the most recent ID, Name, and Pitch Hand 
    
    - First lets create a list of dictionary of all IDs and Names
    

```python
IDsClean = [id for id in data["ID"] if not math.isnan(id)]
dataList = list(data["ID"])
nums = -1
newList = []
for i in range(len(dataList)):
    if math.isnan(dataList[i]):
        newList.append(IDsClean[nums])
    else:
        newList.append(dataList[i])
        nums += 1 
data["ID"] = newList
data
```

```python
NamesClean = [name for name in data["Name"] if type(name) != float]
dataList = list(data["Name"])
nums = -1
newList = []
for name in range(len(dataList)):
    if type(dataList[name]) != float:
        newList.append(dataList[name])
        nums += 1 
    else:
        newList.append(NamesClean[nums])
data["Name"] = newList
data
```

```python
handClean = [hand for hand in data["Pitch Hand"] if type(hand) != float]
dataList = list(data["Pitch Hand"])
nums = -1
newList = []
for name in range(len(dataList)):
    if type(dataList[name]) != float:
        newList.append(dataList[name])
        nums += 1 
    else:
        newList.append(handClean[nums])
data["Pitch Hand"] = newList
data = data.drop(columns = ["Rise", "Tail"])
data
```
